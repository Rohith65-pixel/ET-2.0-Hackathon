```python
import os
PARSED_PATH = os.path.join(os.path.dirname(os.getcwd()),'parsed_files')
dir = os.listdir(PARSED_PATH)[0]
loc = os.path.join(PARSED_PATH,dir)
loc
```

```python
filename = os.path.basename(loc)
md_path = os.path.join(loc,'full.md')
md_path,filename
```

```python
with open(md_path,'rb') as f1:
    content = f1.read()
    with open(os.path.join(os.getcwd(),'input',f'{filename}.md'),'wb') as f2 :
        f2.write(content)
```
